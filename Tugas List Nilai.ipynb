```python
import random

# GENERATE RANDOM
nilaiTugas = []
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))

print('Nilai Tugas adalah: ')
print(nilaiTugas)

nilaiGanjil = []
nilaiLulus = []
nilaiMinimum = 100
nilaiMaksimum = 0
nilaiCount = {}
jumlahNilai = 0
jumlahNilaiKuadrat = 0
# memproses nilai ganjil, nilai > 70, nilai minimun, nilai maksimum
for nilai in nilaiTugas:
    if (nilai%2) != 0:
        nilaiGanjil.append(nilai)
    if nilai > 70:
        nilaiLulus.append(nilai)
    if nilai < nilaiMinimum:
        nilaiMinimum = nilai
    if nilai > nilaiMaksimum:
        nilaiMaksimum = nilai
    nilaiCount[nilai] = 0
    jumlahNilai += nilai
    jumlahNilaiKuadrat += (nilai**2)
    
# memproses modus
# menghitung jumlah setiap nilai
for nilai in nilaiTugas:
    nilaiCount[nilai] += 1
    
# mencari modus
nilaiModus = jumlahModus = 0
listModus = {}
for nilai, jumlah in nilaiCount.items():
    if jumlah > jumlahModus:
        nilaiModus = nilai
        jumlahModus = jumlah
    try:
        listModus[jumlah].append(nilai)
    except KeyError:
        listModus[jumlah] = [nilai]
        
# menghitung standar deviasi
countNilaiTugas = len(nilaiTugas);
nilaiVarian = ((countNilaiTugas * jumlahNilaiKuadrat) - (jumlahNilai ** 2)) / (countNilaiTugas * (countNilaiTugas - 1))
#menghitung akar
def akar(nilai):
    hasil = 0
    while (hasil ** 2) <= nilai:
        hasil += 0.001
    
    return round(hasil, 2)

standarDeviasi = akar(nilaiVarian)

# TAMPILKAN:
# ganjil
print('\nNilai tugas yang ganjil adalah:')
print(nilaiGanjil)
# nilai > 70
print('\nNilai yang lebih besar dari 70 adalah:')
print(nilaiLulus)
# standar deviasi
print('\nStandar deviasi nilai adalah akar dari adalah:')
print(standarDeviasi)
# maksimum
print('\nNilai terendah adalah:')
print(nilaiMinimum)
# minimum
print('\nNilai tertinggi adalah:')
print(nilaiMaksimum)
# modus
print('\nNilai yang paling banyak ada (modus) adalah:', 
      listModus[jumlahModus][0] if len(listModus[jumlahModus]) == 1 else listModus[jumlahModus], 
      '\nSetiap nilai berjumlah', jumlahModus, 'orang')
```
